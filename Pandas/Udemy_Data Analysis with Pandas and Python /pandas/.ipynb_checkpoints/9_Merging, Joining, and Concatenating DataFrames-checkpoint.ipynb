# The pd.concat Method, Part 1 

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
pd.concat(objs = [week1,week2])
pd.concat(objs = [week1,week2], ignore_index = False) ## The original index is still sitting in there
pd.concat(objs = [week1,week2], ignore_index = True) ## ignore original index numbers
```

# The pd.concat Method, Part 2

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
pd.concat(objs = [week1, week2], ignore_index = True)
```

```python
sales = pd.concat(objs = [week1, week2], keys = ['Week 1', 'Week2'])  # not to lose original index
sales.head(3)
```

```python
sales.loc[('Week 1',)]  ## reaching the cells after the key assignment
sales.loc[('Week 1',240)]
sales.loc[('Week 1',240), 'Customer ID']
```

# Outer Joins 

## birlesimlerini gosteriyor. 

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
week1.merge(week2, how = 'outer', on = 'Customer ID', suffixes = [' - Week1', ' - Week2'])
```

```python
merged = week1.merge(week2, how = 'outer', on = 'Customer ID', suffixes = [' - Week1', ' - Week2'], indicator = True)
```

```python
merged['_merge'].value_counts()
```

```python
mask = merged['_merge'].isin(['left_only', 'right_only'])
merged[mask]
```

# Left Joins

## left dataset + right-left intersected

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week1.head(3)
```

```python
foods.head(3)
```

```python
week1.merge(foods,how='left',on = 'Food ID')
```

```python
week1 = week1.merge(foods,how='left',on = 'Food ID', sort = True)  ## sorted according to Food ID
```

```python
week1.head(3)
```

# The left_on and right_on Parameters 

## differently named columns

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week2.head(3)
```

```python
customers.head(3)
```

```python
week2.merge(customers, how = 'left', left_on = 'Customer ID', right_on = 'ID') ## defining the names of columns if different

```

```python
week2.merge(customers, how = 'left', left_on = 'Customer ID', right_on = 'ID').drop('ID', axis = 'columns') ## dropping the duplicated column
```

# Merging by Indexes with the left_index and right_index Parameters 

## index to index merging

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv', index_col = 'ID')  ## Setting the index columns
foods = pd.read_csv('Restaurant - Foods.csv',index_col = 'Food ID') ## Setting the index columns
```

```python
customers.head()
```

```python
foods.head()
```

```python
sales = week1.merge(customers, how = 'left', left_on = 'Customer ID', right_index = True) ## use the right index df as the name of common columns 
sales.head()
```

```python
week1.merge(week2, how='left', left_index= True, right_index = True, suffixes =  [' - Week1', ' - Week 2'])
```

# The .join() Method 

## used when they share the same index

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
satisfaction = pd.read_csv('Restaurant - Week 1 Satisfaction.csv')
```

```python
week1.head()
```

```python
satisfaction.head()
```

```python
week1.merge(satisfaction, how = 'left', left_index = True, right_index = True).head() ## one way to do 
```

```python
week1.join(satisfaction).head() ## another way, but with more versatile and concise 
```
